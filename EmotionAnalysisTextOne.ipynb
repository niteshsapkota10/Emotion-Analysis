```python
results=['anger', 'boredom', 'empty', 'enthusiasm', 'fun', 'happiness',
       'hate', 'love', 'neutral', 'relief', 'sadness', 'surprise',
       'worry']
```

```python
count_vect=CountVectorizer()
count_vect.fit(data['content'].values)
```

```python
print("Print Vocabulary: "+str(count_vect.vocabulary_)+'\n\n')
print("Feature Names: "+str(count_vect.get_feature_names())+'\n\n')
```

```python
ct=count_vect.fit_transform(data['content'].values)
print("The shape of the count is: "+str(ct.shape))
target=data['sentiment_encoded'].values
target
```

```python
trans=TfidfTransformer()
```

```python
train_counts=count_vect.fit_transform(data['content'].values)
train_data=trans.fit_transform(train_counts)
x_train,x_test,y_train,y_test=train_test_split(train_data,target,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
model=mnb.fit(x_train,y_train)
preds=mnb.predict(x_test)
print(accuracy_score(y_test,preds)*100)
print(classification_report(y_test,preds))
print(confusion_matrix(y_test,preds))
```

```python
from sklearn.linear_model import LogisticRegression
rnb=LogisticRegression()
model2=rnb.fit(x_train,y_train)
pred2=rnb.predict(x_test)
print(accuracy_score(y_test,pred2)*100)
print(classification_report(y_test,pred2))
print(confusion_matrix(y_test,pred2))
```

```python
from sklearn.ensemble import RandomForestClassifier
rvc=RandomForestClassifier()
model3=rvc.fit(x_train,y_train)
pred3=rvc.predict(x_test)
print(accuracy_score(y_test,pred3)*100)
print(classification_report(y_test,pred3))
print(confusion_matrix(y_test,pred3))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier()
model4=knc.fit(x_train,y_train)
pred4=knc.predict(x_test)
print(accuracy_score(y_test,pred4)*100)
print(classification_report(y_test,pred4))
print(confusion_matrix(y_test,pred4))
report=classification_report(y_test,pred4)
```

```python
from sklearn.svm import SVC
spm=SVC(kernel='linear',decision_function_shape='ovr')
model5=spm.fit(x_train,y_train)
pred5=spm.predict(x_test)
print(accuracy_score(y_test,pred5)*100)
print(classification_report(y_test,pred5))
print(confusion_matrix(y_test,pred5))
```
